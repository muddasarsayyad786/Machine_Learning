```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
milk = pd.read_csv('Datasets/milk.csv', index_col=0)
milk
```

```python

scaler=StandardScaler().set_output(transform='pandas')
milk_scaled = scaler.fit_transform(milk)
prcom = PCA().set_output(transform='pandas')
pca_data = prcom.fit_transform(milk_scaled)
pca_data.var()
```

#### Percentage of VAriation Explained

```python
(prcom.explained_variance_ / np.sum(prcom.explained_variance_)) *100
```

```python
prcom.explained_variance_ratio_ *100
```

```python
from pca import pca
import matplotlib.pyplot as plt
```

```python
model = pca()
result = model.fit_transform(milk_scaled, col_labels = milk.columns, row_labels=(milk.index))
model.biplot(label=True, legend=True)
for i in np.arange(0, milk.shape[0]):

    model.biplot(pca_data.values[i,0], pca_data.values[i,1], list(milk.index)[i])

plt.show()
```

#### USArrest

```python
us = pd.read_csv('Datasets/USArrests.csv', index_col=0)

scaler=StandardScaler().set_output(transform='pandas')
us_scaled = scaler.fit_transform(us)
prcom = PCA().set_output(transform='pandas')
pca_data = prcom.fit_transform(us_scaled)
pca_data.var()
```

```python
(prcom.explained_variance_ / np.sum(prcom.explained_variance_)) *100
```

```python
np.cumsum((prcom.explained_variance_ / np.sum(prcom.explained_variance_)) *100)
```

```python
model = pca()
result = model.fit_transform(us_scaled, col_labels = us.columns, row_labels=(us.index))
model.biplot(label=True, legend=True)
for i in np.arange(0, us.shape[0]):

    model.biplot(pca_data.values[i,0], pca_data.values[i,1], list(us.index)[i])

plt.show()
```
