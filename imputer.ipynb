```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.impute import SimpleImputer
import pandas as pd
import numpy as np
```

```python
a = np.array([[10, 200, 0.3],
              [9,np.nan, 0.1],
              [8, 240, np.nan],
              [9.5, 225, 0.45]])

pa = pd.DataFrame(a, columns=['X1','X2','X3'])
print("Mean: \n", pa.mean())
```

```python
imputer = SimpleImputer().set_output(transform='pandas')
imputer.fit(pa)
print(imputer.statistics_)
```

```python
pa.mean()
```

```python
imputer.transform(pa)
```
