```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import pandas as pd
import numpy as np
```

#### Standard Scaling

```python
a = np.array([[10, 200, 0.3],
              [9,230, 0.1],
              [9.5, 225, 0.45]])

pa = pd.DataFrame(a, columns=['X1','X2','X3'])
print("Mean: \n", pa.mean())
print('STD: \n ', pa.std())
```

```python
scaler = StandardScaler().set_output(transform='pandas')
scaler.fit(pa)
print(scaler.mean_)
print(scaler.scale_)
```

#### MinMax Scaling

```python
scaler = MinMaxScaler().set_output(transform='pandas')
scaler.fit(pa)
print(scaler.data_min_)
print(scaler.data_min_)
```

```python
scaler.transform(pa)
```
