```python
import pandas as pd
import numpy as np
from sklearn.calibration import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, root_mean_squared_error, mean_absolute_error,confusion_matrix,accuracy_score,classification_report
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
```

```python

wisconsin = pd.read_csv('Cases/Wisconsin/BreastCancer.csv')
le =LabelEncoder()
wisconsin['Class'] = le.fit_transform(wisconsin['Class'])
X , y = wisconsin.drop('Class', axis=1), wisconsin['Class']
X_tain, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=25, stratify=y)
```

```python

lda = LinearDiscriminantAnalysis().set_output(transform='pandas')
lda_trn = lda.fit_transform(X_tain, y_train)
lr = LogisticRegression()
lr.fit(lda_trn, y_train)
lda_tst = lda.transform(X_test)
y_pred = lr.predict(lda_tst)
print(accuracy_score(y_test, y_pred))
```

#### Logistic regression without LDA

```python
vehical = pd.read_csv('Cases/Vehicle_Silhouettes/Vehicle.csv')
X, y = vehical.drop('Class', axis = 1), vehical['Class']
X_tain, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=25, stratify=y)
```

```python
lr = LogisticRegression(max_iter=5000)
lr.fit(X_tain, y_train)
y_pred = lr.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

#### Logistic regression with LDA

```python
lda = LinearDiscriminantAnalysis().set_output(transform='pandas')
lda_trn = lda.fit_transform(X_tain, y_train)
lr = LogisticRegression()
lr.fit(lda_trn, y_train)
lda_tst = lda.transform(X_test)
y_pred = lr.predict(lda_tst)
print(accuracy_score(y_test, y_pred))
```
