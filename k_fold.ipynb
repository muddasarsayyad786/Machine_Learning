```python
from sklearn.tree import DecisionTreeClassifier
from tqdm import tqdm
from sklearn.model_selection import KFold, cross_val_score,StratifiedKFold


depts = [None, 3, 4, 5, 6, 7]
min_sample = [2, 10, 0.005, 0.01, 0.05, 0.1]
min_leaf = [1, 10, 0.025, 0.01, 0.05, 0.1]
scores = []

for d in tqdm(depts):
    for s in min_sample:
        for f in min_leaf:
            dtc = DecisionTreeClassifier(random_state=24, 
                                         max_depth=d, 
                                         min_samples_split=s, 
                                         min_samples_leaf=f)
            
            kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=25)
            result = cross_val_score(dtc, X, y, scoring='f1_macro', cv=kfold)
            
            scores.append([d, s, f, np.mean(result)])

df_scores = pd.DataFrame(scores, columns=['max_depth', 'min_samples_split', 'min_samples_leaf', 'score'])
df_scores.sort_values(by='score', ascending=False)
```
