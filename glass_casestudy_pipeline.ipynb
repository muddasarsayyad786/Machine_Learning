```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score, root_mean_squared_error, mean_absolute_error,confusion_matrix,accuracy_score,classification_report
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
import os
from sklearn.metrics import log_loss
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
```

```python
glass = pd.read_csv('Cases\Glass_Identification\Glass.csv')
glass.head()

X,y = glass.drop('Type', axis=1), glass['Type']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30, random_state=25, stratify=y)



mm_scaler = MinMaxScaler()
std_scaler = StandardScaler()
ks = np.arange(1,16)
scores = []

for k in ks:
    for s in [mm_scaler,std_scaler]:
        knn = KNeighborsClassifier(n_neighbors=k)
        pipe = Pipeline(steps=[('SCI', s),('KNN',knn)])
        pipe.fit(X_train, y_train)
        y_pred_proba = pipe.predict_proba(X_test)
        scores.append([k,s,log_loss(y_test, y_pred_proba)])

df_scores = pd.DataFrame(scores, columns=['k', 'S', 'log_loss'])
df_scores.sort_values(by='log_loss')
```

```python
#apply best model
knn = KNeighborsClassifier(n_neighbors=9)
pipe = Pipeline(steps=[('SCI', mm_scaler), ('KNN', knn)])
pipe.fit(X, y)
```
