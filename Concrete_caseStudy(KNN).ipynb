```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, root_mean_squared_error, mean_absolute_error,confusion_matrix,accuracy_score,classification_report
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
import os
```

```python
df = pd.read_csv('Cases\Concrete_Strength\Concrete_Data.csv')
df.head()
```

```python
X,y = df.drop('Strength', axis=1), df['Strength']
```

```python
#Without Scaling
X_tain, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30)
model = KNeighborsRegressor()
model.fit(X_tain, y_train)
y_pred = model.predict(X_test)
error_score = mean_absolute_error(y_test,y_pred)
print(error_score)
```

```python
#With Scaling
scaler = StandardScaler().set_output(transform='pandas')
X_train_scaled = scaler.fit_transform(X_tain)
X_test_scaled = scaler.fit_transform(X_test)
# y_test_scaled = scaler.fit_transform(y_test)
# y_train_scaled = scaler.fit_transform(y_train)

model = KNeighborsRegressor()
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)
error_score = mean_absolute_error(y_test,y_pred)
print(error_score)
```
