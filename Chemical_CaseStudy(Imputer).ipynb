```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, root_mean_squared_error, mean_absolute_error,confusion_matrix,accuracy_score,classification_report
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
import os
```

```python
chem = pd.read_csv('Cases\Chemical Process Data\ChemicalProcess.csv')


chem.head()
```

```python
X,y = chem.drop('Yield', axis=1), chem['Yield']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.30,random_state=25)
```

Mean Strategy

```python
imput = SimpleImputer(strategy='mean')
scale = StandardScaler()
```

```python
#Without Scaling
X_trn_imp = imput.fit_transform(X_train)
X_tst_imp = imput.transform(X_test)
ks = np.arange(1,16)
score = []

for k in ks:
    model = KNeighborsRegressor(n_neighbors=k)
    model.fit(X_trn_imp, y_train)
    y_pred = model.predict( X_tst_imp )
    score.append([k, mean_absolute_error(y_test, y_pred)])

df_score = pd.DataFrame(score, columns=["K", "score"])
df_score.sort_values(by='score', ascending=True)
```

```python
#With Scaling
X_trn_imp = imput.fit_transform(X_train)
X_tst_imp = imput.transform(X_test)
X_trn_imp_scal = scale.fit_transform(X_trn_imp)
X_tst_imp_scal = scale.transform(X_tst_imp)
ks = np.arange(1,16)
score = []

for k in ks:
    model = KNeighborsRegressor(n_neighbors=k)
    model.fit(X_trn_imp_scal, y_train)
    y_pred = model.predict( X_tst_imp_scal )
    score.append([k, mean_absolute_error(y_test, y_pred)])

df_score = pd.DataFrame(score, columns=["K", "score"])
df_score.sort_values(by='score', ascending=True)
```

```python
imputers = ['mean', 'median']
scaler =StandardScaler()
ks = np.arange(1,16)
scores = []
for i in imputers:

    for k in ks:
        imput = SimpleImputer(strategy=i)
        model = KNeighborsRegressor(n_neighbors=k)
        pipe = Pipeline(steps=[('IMP',imput), ('SCL', scaler), ('Model', model)])
        pipe.fit(X_train, y_train)
        y_pred = pipe.predict( X_test )
        scores.append([i, k, mean_absolute_error(y_test,y_pred)])

df_score = pd.DataFrame(scores, columns=['impute', 'k', 'score'])
df_score.sort_values(by='score')
```
